# Boosting

```python
-->The main thing in boosting it is sequintal in nature.
-->Diff b/w bagging and boosting is that bagging is parallel in nature but boosting is sequintal in nature.
-->In boosting there are three algorithms :
    
    1)Adaboost : stump means decision tree with hight = 1
    
    2)Gradient Boosting 
    
    3)Xgboost
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv('bank.csv')
df.head()
```

```python
df.shape
```

```python
df.deposit.value_counts()
```

```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

# Boosting Classifiers

# Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf=AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=200)
```

```python
ada_clf.fit(xtrain,ytrain)
```

```python
y_pred=ada_clf.predict(xtest)
```

```python
accuracy_score(ytest, y_pred)
```

```python
print(classification_report(ytest,y_pred))
```

# GradientBoosting

```python
from sklearn.ensemble import GradientBoostingClassifier
import time
```

```python
gbc_clf=GradientBoostingClassifier()
```

```python
gbc_clf.fit(xtrain,ytrain)
```

```python
start=time.time()
gdc_clf.fit(xtrain,ytrain)
end=time.time()
time_elapsed=end - start
print(time_elapsed)
```

```python
y_pred=gdc_clf.predict(xtest)
print('Accuracy : ',accuracy_score(ytest,y_pred))
```

```python
print(classification_report(ytest,y_pred))
```

```python
params={'n_estimators' : [10,80,100,120],
        'learning_rate' : [0.1,0.3,1.0],   #rate at which model is learning, 0 to 1
        'max_depth' : [3,4,5]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grd_gbc_clf=GridSearchCV(gbc_clf,params)
```

```python
grd_gbc_clf.fit(xtrain,ytrain)
```

```python
grd_gbc_clf.best_params_
```

```python
pred_using_GS=grd_gbc_clf.predict(xtest)
```

```python
print('Accuracy : ',accuracy_score(pred_using_GS,ytest))
# print('Estimators : ',grd_gbc_clf.n_estimators_)
```

```python
Difference Between Bagging & Boosting

-->The models are created independently.

-->The model creation is dependent on the previous ones.

-->In Bagging, training data subsets are drawn randomly with a replacement for the training
dataset.

-->In Boosting, every new subset comprises the elements that were misclassified by
previous models.

-->Base classifiers are trained parallelly.

-->Base classifiers are trained sequentially
```
