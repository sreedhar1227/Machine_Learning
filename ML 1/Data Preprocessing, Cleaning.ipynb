# Feature Scaling


```python
df.head()
```

```python
#standarlization formula   "(x-u(mean)/sigma)"

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
df[['age', 'trtbps', 'chol', 'thalachh']]=scale.fit_transform(df[['age', 'trtbps', 'chol', 'thalachh']])
```

```python
df.head()
```

```python
#Normalization formula "(x-max(x)/max(x)-min(x))"

from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
df[['age', 'trtbps', 'chol', 'thalachh']]=scale.fit_transform(df[['age', 'trtbps', 'chol', 'thalachh']])
```

```python
df.head()
```

# Data Preparation

```python
from sklearn.model_selection import train_test_split
```

```python
#Split dataset in features and target variable

feature_cols=['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall', 'output']
x = df[feature_cols] # features
y = df.output # target variables
```

```python
#Split dataset into training set and test set

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=9) #80% training & 20% test
```

# Supervised Learning

```python
->In supervised learning there are two catagerious 1)Classification 2)Regression

->To identify "Regression problem" ==> where we can predict the values of anything like house rent, pertol,.... 
        is called regression problems. "output variable will be continous in nature". output variable value is not fixed.
    

->To identify "Classification problem" ==>"output will be categrical in nature" like in the form of "true or false", "yes or no"
                                          Binary class classification (0 & 1)
                                          Digit class classification (0,1,2...9)
        
TYPES OF REGRESSION ALGORITHM :
    
1)Linear regression : It is used only when there is one independent variable

2)Multiple linear regression : It is used only when there is more than one independent variable

3)Polynomial regression

4)Ridge regression

5)Lasso regression

6)ElasticNet regression
```
